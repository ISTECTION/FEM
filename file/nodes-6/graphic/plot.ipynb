````python
import matplotlib.pyplot as plt
````

````python
X = [2, 2, 3, 7, 2, 3, 2, 2]
Y = [1, 0, 1, 4, 4, 1, 1, 4]

plt.plot(X, Y)
plt.xlim(1, 7.25)
plt.ylim(-0.1, 5.25)

plt.text(1.75, 0.05, "0", fontsize=14)
plt.text(1.75, 1.05, "1", fontsize=14)
plt.text(3.05, 0.75, "2", fontsize=14)
plt.text(1.75, 4.05, "3", fontsize=14)
plt.text(7.05, 4.05, "4", fontsize=14)

plt.show()
````
